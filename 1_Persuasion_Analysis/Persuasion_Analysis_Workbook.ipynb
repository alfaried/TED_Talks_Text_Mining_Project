# 1.1 Persuasion Analysis

```python
from empath import Empath
lexicon = Empath()
```

```python
lexicon.create_category("persuasion",["anger","anticipation","disgust","fear","joy","negative","positive","sadness","surprise","trust"])
```

```python
import pandas as pd

df = pd.read_csv("../Processed_Dataset/cleaned_dataset.csv")
```

```python
transcripts = df["transcript"]
```

```python
for i,script in enumerate(transcripts):
    sentences = script.split(".")
    for sentence in sentences:
        lexi_dict = lexicon.analyze(sentence,categories=["persuasion"], normalize=True)
        if lexi_dict!= None and lexi_dict["persuasion"] > 0.0:
            print(lexi_dict)
            print(sentence)
```

```python
from empath import Empath
lexicon = Empath()
```

```python
import pandas as pd
import xlwt

cleaned_transcripts = pd.read_csv('./Processed_Dataset/cleaned_dataset.csv', encoding="ISO-8859-1")

cleaned_transcripts.head()
```

```python
transcripts_list = cleaned_transcripts["transcript"].tolist()
```

```python
# SAD EMotion: sadness,
#
for i,transcript in enumerate(transcripts_list):
    str_list = transcript.split(".")
    for sentence in str_list:
        emotion_score = lexicon.analyze(sentence,categories=["zest", "sadness"])
        if emotion_score["zest"] or emotion_score["sadness"] > 0.0:
            print(sentence)
    if i == 1:
        break
```

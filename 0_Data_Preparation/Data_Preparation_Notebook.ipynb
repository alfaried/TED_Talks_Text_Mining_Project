Merge ted_main.csv and transcripts.csv datasets on URL column

```python
import pandas as pd
import xlwt

meta_df = pd.read_csv('../Raw_Dataset/ted_main.csv', encoding="ISO-8859-1")
transcripts_df = pd.read_csv('../Raw_Dataset/transcripts.csv', encoding="ISO-8859-1")
all_df = pd.merge(meta_df, transcripts_df, how='left', left_on='url', right_on='url')

all_df.to_csv('../Processed_Dataset/all.csv', encoding="ISO-8859-1")
```
